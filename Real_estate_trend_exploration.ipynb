```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import plotly.express as px
```

```python
# Load data
dir = 'data/'
inventory = pd.read_csv(dir+'Inventory_forsale_raw_sfr_weekly.csv')
```

```python
#inventory[inventory.StateName == 'CA']
regions = [395078, 395057] # US 102001, Seattle, Chicago 394463, SF
dates = [c for c in inventory.columns if c.startswith('20')]
```

```python
df = inventory[inventory.RegionID.isin(regions)]
df = pd.melt(df, id_vars = 'RegionName', value_vars = dates, var_name = 'week', value_name = 'inventory')
```

```python
fig = px.line(df, x="week", y="inventory", color = 'RegionName')
fig.show()
```
