```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import plotly.express as px
```

```python
# Load data
dir = 'data/'
inventory = pd.read_csv(dir+'Inventory_forsale_raw_sfr_weekly.csv')
sales = pd.read_csv(dir + 'sales_count_raw_sfrcondo_month.csv')
sale_price = pd.read_csv(dir + 'median_sale_price_raw_sfr_week.csv')
rent = pd.read_csv(dir + 'ZORentIndex_AllHomesPlusMultifamily_smooth_seasonadjusted.csv')
```

```python
#inventory[inventory.StateName == 'CA']
regions = [395078] # US 102001, Seattle, Chicago 394463, SF 395057
```

```python
df = inventory[inventory.RegionID.isin(regions)]
dates = [c for c in inventory.columns if c.startswith('20')]
df = pd.melt(df, id_vars = 'RegionName', value_vars = dates, var_name = 'week', value_name = 'inventory')
fig = px.line(df, x="week", y="inventory", color = 'RegionName')
fig.show()
```

```python
df = sales[sales.RegionID.isin(regions)]
dates = [c for c in sales.columns if c.startswith('20')]
df = pd.melt(df, id_vars = 'RegionName', value_vars = dates, var_name = 'week', value_name = 'sales')
fig = px.line(df, x = 'week', y = 'sales')
fig.show()
```

```python
df = sale_price[sale_price.RegionID.isin(regions)]
dates = [c for c in sale_price.columns if c.startswith('20')]
df = pd.melt(df, id_vars = 'RegionName', value_vars = dates, var_name = 'week', value_name = 'median_price')
fig = px.line(df, x = 'week', y = 'median_price')
fig.show()
```

```python
df = rent[rent.RegionID.isin(regions)]
dates = [c for c in rent.columns if c.startswith('20')]
df = pd.melt(df, id_vars = 'RegionName', value_vars = dates, var_name = 'week', value_name = 'rent')
fig = px.line(df, x = 'week', y = 'rent')
fig.show()
```
